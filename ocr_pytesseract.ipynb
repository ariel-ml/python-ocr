## Trained Tesseract MRZ reader

```python
import os
import pytesseract
import cv2
import utils
```

```python
# set tessdata prefix to models folder
os.environ["TESSDATA_PREFIX"] = "models"

# set white list chars to config
white_list = "ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789<"

tess_config = f"tessedit_char_whitelist={white_list}"
```

```python
image = cv2.imread("temp/roi.png")
```

```python
utils.display(image)
```

```python
mrz_code = pytesseract.image_to_string(image, lang="mrz", config=tess_config)
```

```python
print(mrz_code)
```
